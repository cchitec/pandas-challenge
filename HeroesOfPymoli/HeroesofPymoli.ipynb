```python
# Import Dependencies
import pandas as pd
```

```python
#File to Load 
csv_path = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(csv_path)
purchase_data_df = pd.DataFrame(purchase_data)
purchase_data_df.head(30)
```

```python
# Display the total number of players
total_players = purchase_data.loc[:, ["Gender", "SN", "Age"]]
total_players = total_players.drop_duplicates()
total_players = total_players.count()[0]# Display the total number of players
pd.DataFrame({"Total Players": [total_players]})
```

```python
#Purchase analysis 

#number of unique items
unique_item = len(pd.unique(purchase_data_df['Item Name']))
#average purchase price
average_price = purchase_data_df["Price"].mean()
#Total Number of Purchases
total_purchase = len(purchase_data_df.groupby("Purchase ID")["SN"].count())
#Total revenue
total_revenue = purchase_data_df["Price"].sum()

#purchase_data_df.dtypes

purchasing_analysis_df = pd.DataFrame([{"Number of Unique Items" : unique_item, "Average Price" : average_price, "Number of Purchases" : total_purchase, "Total Revenue" : total_revenue}])
purchasing_analysis_df
```

```python
#Gender Demographics
#Total number of players of each gender
total= purchase_data[["Gender", "SN"]].drop_duplicates().Gender.value_counts()


#Percentage of players
percent = purchase_data[["Gender", "SN"]].drop_duplicates().Gender.value_counts(normalize=True).astype(float).map("{:.2%}".format)

gender_demographics_df = pd.DataFrame({"Total Count" : total,   "Percentage of Players" : percent})
gender_demographics_df
```

```python
#Purchasing Gender Analysis
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

#purchase count
pur_count = purchase_data_df["Gender"].value_counts()

#average purchase price
gender_average = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")
#total revenue by gender
total_revenue = purchase_data_df.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")

#average price per person by gender
avg_price_per = total_revenue / pur_count

gender_purchase_df = pd.DataFrame({
    "Purchase Count" : pur_count,
    "Average Purchase Price" : gender_average,
    "Total Purchase Value" : total_revenue,
    "Avg Total Purchase per Person" :  avg_price_per
})
gender_purchase_df
```
